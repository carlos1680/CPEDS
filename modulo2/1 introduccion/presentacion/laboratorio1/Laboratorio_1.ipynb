## Bucles (for y while)

### Bucle For

```python
for i in range(0,6):
  print(f"Indice: {i}, ejecucion: {i+1}, iMenor3: {i < 3}")
```

```python
estudiantes = ["José", "Luis", "Marcelo", "Sebastián", "Camila"]

print(len(estudiantes))
for i in range(len(estudiantes)):
  print(f"Indice: {i}, nombre:{estudiantes[i]}")
```

```python
for i in estudiantes:
  print(i)
```

```python
numeros = [4, 5, 2, 12]

for i in numeros:
  print(i + 5)
```

### Bucle while

```python
n = 0

while n < 5:
  print(n)
  n += 1
```

```python
# Bucle infinito
# Atajo para cotar ejecución: Ctrl + M + I

n = 0

while n < 5:
  print(n)
  n = n - 1
```

## Clases

```python
class Docente:

  def __init__(self, nombre, materias, edad): # El self hace referencia a que la variable modificada es la correspondiente a la clase únicamente.
    self.nombre = nombre
    self.materias = materias
    self.edad = edad

  def cumple_anios(self):
    self.edad += 1

  def agregar_materias(self, materia_nueva): # El self como primer parámetro permite llamar a la función solo si está la clase
    self.materias.append(materia_nueva)
```

```python
Daniel = Docente(
    nombre = "Daniel Blanc",
    materias = ["Excel", "Python", "Power BI"],
    edad = 29
)
```

```python
print(Daniel.edad)
Daniel.cumple_anios()
print(Daniel.edad)
```

```python
print(Daniel.materias)
Daniel.agregar_materias("R")
print(Daniel.materias)
```
