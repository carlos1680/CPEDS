### loc (basado en etiquetas)

```python
nba.loc["Utah Jazz"] # El primer elemento hace referencia al índice. 
```

```python
nba.loc["Chicago Bulls" : "Detroit Pistons"]
```

```python
nba.loc[: "Chicago Bulls"]
```

```python
nba.loc[["Chicago Bulls", "Detroit Pistons"]]
```

Consideraciones de la selección:
*  Cuando son dos elementos continuos, se puede aplicar un slicing directamente: elemento_1 : elemento_2
*  Cuando son dos elementos que no son continuos, se les debe pasar en una lista: [elemento_1, elemento_2, elemento_3]

```python
nba.loc[["Chicago Bulls", "Detroit Pistons"], "Name" : "Position"]
```

El loc acepta condiciones booleanas para su selección

```python
nba.loc[(nba.Team == 'Chicago Bulls') | (nba.Age <25)] # las condiciones deben estar entre parentesis # | operador OR
```

### Función where

```python
nba.where(nba["Salary"] < 5000000, other = 0)
# Cuando son menores deja sin toca, en caso contrario cambia el valor por 0
```

### Función query

```python
nba.query('Salary < 5000000')
```

```python
condicion = nba["Team"].isin(["Boston Celtics", "Chicago Bulls", "Utah Jazz"]) # Para buscar una cierta cantidad de elementos
nba[condicion]
```

```python
condicion = nba["Team"].isin(["Boston Celtics", "Chicago Bulls", "Utah Jazz"]) # En caso de querer filtrar por la negativa, ponemos el símbolo de equivalente adelante
nba[~condicion]
```

```python
# Para simular el 'like' de sql, se puede utilizar la función contains.
# Hay que aclarar qué se busca, qué se hace con los NA, y si es sensible a las mayúsculas
# Si hay valores faltantes en la serie "Team", se tratarán como si no contuvieran la cadena "boston"
# La búsqueda no será sensible a mayúsculas y minúsculas
condicion = nba["Team"].str.contains("boston", na = False, case = False)
nba[condicion]
```

### Valores duplicados

```python
import pandas as pd
```

```python
nba.tail(10)
```

```python
nba.duplicated(subset=["Age", "Team"],
               keep = 'last').tail(10)
```

keep='last': Este argumento especifica que la última ocurrencia de cada conjunto de duplicados se considerará como no duplicada, mientras que las anteriores se considerarán duplicadas

```python
nba[~nba["Age"].duplicated(keep = 'first')].head()
```

La primera ocurrencia de un valor se considera no duplicada, mientras que las subsiguientes se consideran duplicadas.
La sentencia selecciona las filas del DataFrame nba donde la edad es única o la primera ocurrencia de cada valor de edad y muestra las primeras filas de ese resultado.
