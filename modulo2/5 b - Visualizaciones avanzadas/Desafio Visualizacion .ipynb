```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
tips.shape
```

```python
tips.info()
```

```python
tips["day"].value_counts()
```

```python
tips["day"].value_counts().plot(kind="bar")
plt.title("Día preferido para comer")
plt.xlabel("Día")
plt.legend("Cantidad")
plt.figure(figsize=(6,3))
```

```python
plt.figure(figsize=(6,3))
sns.histplot(data=tips,x=("day"))
plt.title("Día preferido para comer")
plt.xlabel("Día")
plt.ylabel("Comensales")
```

```python
plt.figure(figsize=(8,4))
sns.barplot(data=tips,x=("day"),y=("total_bill"),hue="sex", errorbar=None, palette="viridis")
plt.title("Pagos distribuidos por género")
plt.xlabel("Día")
plt.ylabel("Monto")
plt.legend(loc="upper left")
```

```python
plt.figure(figsize=(8,4))
sns.lmplot(data=tips,x= "size", y=("total_bill"),col="smoker")
plt.title("Pagos distribuidos por género")
plt.xlabel("Día")
plt.ylabel("Monto")
plt.legend(loc="upper left")
```

```python
sns.set_palette("Spectral")
```

```python
sns.relplot(
    data=tips,
    x="day", y="total_bill",
    col="smoker",
    kind="line",
    height=5, aspect=.75, facet_kws=dict(sharex=False),
)
```

```python
plt.figure(figsize=(8,4))
sns.lineplot(data=tips,x= "size", y=("total_bill"),hue="smoker")
plt.title("Comparación de monto entre fumadores y no fumadores")
plt.xlabel("Día")
plt.ylabel("Monto")
plt.legend(loc="lower right")
```

```python
sns.scatterplot(data=tips,x="total_bill",y="tip")
plt.title ("Propina vs Monto")
plt.xlabel("Monto")
plt.ylabel("Propina")
```

```python
tips["time"].unique()
```

```python
plt.figure(figsize=(8,4))
sns.barplot(data=tips,x=("day"),y=("total_bill"),hue="time", palette="deep")
plt.title("Distribución del monto por tiempo de comida")
plt.xlabel("Día")
plt.ylabel("Monto")
plt.legend(loc="upper left")
```

```python
plt.figure(figsize=(8,4))
sns.barplot(data=tips,x=("tip"),y=("total_bill"),hue="sex", palette="deep")
plt.title("Distribución del monto por tiempo de comida")
plt.xlabel("Día")
plt.ylabel("Monto")
plt.legend(loc="upper left")
```
