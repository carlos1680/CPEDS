##### Función agg()

```python
sectors3 = fortune.groupby('Sector')
```

**Aplicación de diferentes funciones de agregación a columnas específicas**

```python
sectors3.agg({
    "Revenue" : "sum",
    "Profits" : "sum",
    "Employees" : "mean"
})
```

```python
sectors.agg(Maximo =("Revenue", 'max'), Min_revenue =("Revenue", 'min'), Mean_revenue =("Revenue", 'mean'))
#podemos colocarle el nombre que queramos
```

```python
sectors.agg({
    "Revenue" : "sum",
    "Profits" : ["sum", "mean"],
    "Employees" : ["size", "sum"]
})
```

## Función pivot_table()

```python
import numpy as np
mean_employees = fortune["Employees"].mean() # calculo el promedio 
fortune["Company_size"] = np.where(fortune["Employees"] > mean_employees, "Big", "Small")
```

```python
fortune["Company_size"].value_counts()
```

```python
pd.pivot_table(data = fortune, #dataframe
               index = ['Sector'],
               columns = ['Company_size'],
               aggfunc = 'size') # 'size' indica que se está contando el número de ocurrencias (tamaño) 
```

```python
pd.pivot_table(data = fortune,
               index = ['Industry'],
               columns = ['Company_size'],
               aggfunc = 'size' 
               fill_value = 0)
```

```python
pd.pivot_table(data = fortune,
               values = 'Revenue', #columna numerica 
               index = ['Sector','Industry'],
               columns = ['Company_size'],
               aggfunc = ["sum", 'size'], # suma se calcula sobre el revenue y la cuenta sobre el tamaño de la compania 
               fill_value = 0)
```

## Función corr()

```python
fortune.corr(method = 'pearson', numeric_only = True)
```

El coeficiente de correlacion es una métrica que explica cómo de asociadas están dos variables, va del -1 al 1. Uno significa totalmente correlacionadas positivamente, 0 significa nada correlacionadas y -1 significa totalmente correlacionadas negativamente. Normalmente, correlaciones menores del 0.5 (positivo o negativo) se consideran correlaciones muy bajas. Correlaciones entre 0.5 y 0.8 son correlaciones moderadas y no permiten tomar decisiones muy claras y correlaciones por encima del 0.8 son altas y sí que implican una clara correlación entre ambas variables.
