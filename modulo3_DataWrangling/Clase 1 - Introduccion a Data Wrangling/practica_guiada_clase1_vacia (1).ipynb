## 4 - Iterar el dataframe fila a fila (Caso #1)

```python
inicio = time.time()
direcciones = []
for ind,fila in df.iterrows():
    direcciones.append(fila['domicilio_fiscal'])
fin = time.time()

tiempo_proceso = fin - inicio

print(f'El proceso llevo {str(tiempo_proceso)}')
```

## 5 - Iterar el dataframe por el indice (Caso #2)

```python
inicio = time.time()
temp = []
for idx in range(0, len(df)):
  temp.append(df['domicilio_fiscal'].iloc[idx])
fin = time.time()
diff3 = fin - inicio
print("La diferencia: " + str(diff3))
tiempos = pd.concat([tiempos, pd.DataFrame(
    {'Tipo_Test': ['Iterar por indice'], 'Tiempo': [diff3]})], ignore_index=True)
tiempos
```

## 6 - Iterar desde un diccionario (Caso #3)

```python
df_dict = df.to_dict('records')
inicio = time.time()
temp = []
for row in df_dict:
  temp.append(row['domicilio_fiscal'])
fin = time.time()
diff4 = fin - inicio
print("La diferencia: " + str(diff4))
# resultado = resultado.append({'Tipo_Test':'Iterar desde un diccionario','Tiempo':diff},ignore_index=1)
tiempos = pd.concat([tiempos, pd.DataFrame(
    {'Tipo_Test': ['Iterar desde un diccionario'], 'Tiempo': [diff4]})], ignore_index=True)
tiempos
```

## 7 - Iterar con una funcion lambda (Caso #4)

```python
df_dict = df.to_dict('records')
inicio = time.time()
temp = []
temp.append(df['domicilio_fiscal'].apply(lambda x: x))
fin = time.time()
diff5 = fin - inicio
print("La diferencia: " + str(diff5))
# resultado = resultado.append({'Tipo_Test':'Iterar con funcion lambda','Tiempo':diff},ignore_index=1)
tiempos = pd.concat([tiempos, pd.DataFrame(
    {'Tipo_Test': ['Iterar con funcion lambda'], 'Tiempo': [diff5]})], ignore_index=True)
```
