## 3.4 Asimetría

```python
# Vamos a calcular la asimetría usando la formula completa
x = [8.0, 1, 2.5, 4, 28.0]
n = len(x)
mean_ = sum(x) / n
var_ = sum((item - mean_)**2 for item in x) / (n - 1)
std_ = var_ ** 0.5
skew_ = (sum((item - mean_)**3 for item in x) * n / ((n - 1) * (n - 2) * std_**3))
skew_
```

Vemos que el resultado es positivo, por lo tanto x tiene una cola a derecha.

```python
# Ahora vamos a usar la funcion especifica
scipy.stats.skew(x)
```

# 4. Medidas de relación entre variables




## 4.1 Correlación

```python
# Vamos a generar dos variables para analizar luego
x = list(range(-10, 11))
y = [0, 2, 2, 2, 2, 3, 3, 6, 7, 4, 7, 6, 6, 9, 4, 5, 5, 10, 11, 12, 14]
x_, y_ = np.array(x), np.array(y)
x__, y__ = pd.Series(x_), pd.Series(y_)
```

### 4.1.1 Covarianza

```python
# Calculamos la covarianza usando la fórmula
n = len(x)
mean_x, mean_y = sum(x) / n, sum(y) / n
cov_xy = (sum((x[k] - mean_x) * (y[k] - mean_y) for k in range(n)) / (n - 1))
cov_xy
```

Vamos a calcular la **matriz de covarianza** entre variables. Aquí vemos la covarianza de todas las variables con todas las variables.

```python
cov_matrix = np.cov(x_, y_)
cov_matrix
```

Vemos que la covarianza que nos interesa aparece en la esquina superior derecha (`cov_matrix[0, 1]`) y en la esquina inferior izquierda (`cov_matrix[1, 0]`) ya que se corresponden ambas a la covarianza entre x e y. 

La covarianza de la esquina superiror izquierda (`cov_matrix[0, 0]`) corresponde a la covarianza entre x y x y la inferior derecha (`cov_matrix[1, 1]`) a la covarianza entre y e y. Notese que la covarianza entre una variabe y ella misma es en definitiva la varianza de la variable.

```python
# Podemos calcular directamente las covarianzas entre dos variables
cov_xy = x__.cov(y__)
print(cov_xy)
cov_xy = y__.cov(x__)
print(cov_xy)
```

### 4.1.2 Coeficiente de correlación

```python
# Vamos a calcularlo usando la fórmula
var_x = sum((item - mean_x)**2 for item in x) / (n - 1)
var_y = sum((item - mean_y)**2 for item in y) / (n - 1)
std_x, std_y = var_x ** 0.5, var_y ** 0.5
r = cov_xy / (std_x * std_y)
r
```

```python
# Vamos a usar la funcion de scipy
r, p = scipy.stats.pearsonr(x_, y_)
print(r)
print(p)
```

Esta función nos devuelve el coeficiente de correlación (*r*) junto con su p-valor (*p*).

Numpy tiene una matriz de correlación que simplemente nos devuelve los coeficientes de correlación. Su interpretación es análoga a la de la matriz de covarianzas.


```python
corr_matrix = np.corrcoef(x_, y_)
corr_matrix
```

También podemos usar la función de Pandas para 

```python
(x_, y_).corr
```

También podríamos usar una regresión lineal para calcular el coeficiente de correlación.

```python
result = scipy.stats.linregress(x_, y_)
r = result.rvalue
r
```
