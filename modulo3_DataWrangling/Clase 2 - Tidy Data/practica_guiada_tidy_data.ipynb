# 2. Preguntas específicas

Para contestar preguntas especificas: realizar filtros y selecciones de columnas existen dos tipos de métodos:

* **Método basado en indexing**: `DataFrame.loc[selección filas, selección columnas]`

* **Método basado en expresiones**: `DataFrame.query(expresión)`

Queremos hacer un análisis sobre estos datos con las siguientes preguntas:

a. ¿Cuáles géneros de música hay?

b. ¿Cuáles son las canciones de género Electronica? 

c. ¿Cuáles son los/las artistas y canciones del género Rock que ingresaron al ranking a partir de octubre? *italicized text*

### a. ¿Cuáles géneros de música hay?

Para contestar esta pregunta podemos usar el método `DataFrame.column.unique()` o la funcion `pd.unique()` que nos permiten obtener los valores únicos de una columna

```python
# Vemos los generos distintos que existen
pd.unique(billboard.genre)
```

```python
# La alternativa
billboard.genre.unique()
```

### b. ¿Cuáles son las canciones de género Electronica? 
Veamos los dos métodos para contestar esta pregunta

```python
# Metodo basado en indexing
## Ponemos la condicion del filtro
billboard.loc[billboard.genre=='Electronica']
```

```python
# Para seleccionar solo la cancion lo añadimos en el segundo argumento del .loc[]
billboard.loc[billboard.genre=='Electronica', 'track']
```

```python
# Metodo basado en expresiones
## Definimos el filtro mediante la expresion
billboard.query("genre=='Electronica'")
```

El **método basado en expresiones** puede ser más sencillo de leer pero no se puede seleccionar columnas directamente mientras que el **método basado en indexing** puede ser más complejo de leer pero permite seleccionar filas y columnas directamente

```python
# Luego de utilizar query podemos seleccionar las columnas
billboard.query("genre=='Electronica'")['track']
```

### c. ¿Cuáles son los/las artistas y canciones del género Rock que ingresaron al ranking a partir de octubre? 

Para contestar esta pregunta debemos filtrar por el genero y la fecha de ingreso y quedarnos con las columnas de artista y canción

Para indicar que se deben cumplir ambas condiciones utilizamos el simbolo `&` entre ambos filtros: (condicion 1) & (condicion 2). Esto significa que se deben cumplir ambas condiciones en simultaneo

```python
# Metodo basado en indexing
billboard.loc[(billboard.genre=='Rock') & (billboard['date.entered'] >= '2000-10-01'), ['artist.inverted','track']]
```

# 4. Tidy data

Recordemos que las condiciones para que un dataset se considere ordenado (tidy) son:

* Cada variable forma una columna
* Cada observación forma una fila
* Cada tipo de unidad observacional forma una tabla

¿El dataset de billboard cumple con estas condiciones?

No. Observamos que:

1. La variable del número de semana en el ranking no está como una columna
2. Aunque es discutible, podríamos argumentar que hay dos unidades observaciones en la misma tabla: características de la canción y posición en el ranking

Para resolver el primer problema vamos a utilizar la función `pd.melt()`

```python
# Al ejecutar esta celda se abre la documentación de la función
pd.melt?
```

Nuestro primer paso va a ser especificar las **id_vars**, variables identificadoras y las **value_vars**.

¿Cuáles son cada una de ellas?

**id_vars**: "year", "artist.inverted", "track", "time", "genre", "date.entered", "date.peaked"

Aunque si tenemos en cuenta la pregunta puntual pueden ser menos

**value_vars**: todas las variables de las semanas

Tengamos en cuenta que no es necesario poner la lista de todas las variables ya que la función considera como value_vars a todas aquellas columnas que no aparecen en id_vars

```python
# Aplicamos la función melt y tomamos una muestra
pd.melt(billboard,
        id_vars=["year", "artist.inverted", "track", "time", "genre", "date.entered", "date.peaked"]).sample(5)
```

Vemos que ahora tenemos el número de la semana como una columna y el ranking en esa semana como otra.

Pero, podemos mejorar el resultado y que nuestras nuevas columnas no tengan nombres genericos como "variable" y "value". Esto lo hacemos utilizando los argumentos **var_name** y **value_name**

```python
# Creamos nuestro dataset tidy
billboard_tidy = pd.melt(billboard, 
                      id_vars=["year", "artist.inverted", "track", "time", "genre", "date.entered", "date.peaked"], 
                      var_name="week", 
                      value_name="ranking")
```

```python
billboard_tidy.head()
```

```python
# Dimensiones del dataset
billboard_tidy.shape
```

Ahora el dataset tiene 24092 filas y 9 columnas en comparación a las 317 filas y 83 columnas originales.

Si bien podemos realizar algunas transformaciones adicionales para mejorar el dataset, son necesarias técnicas que veremos más adelante en el curso. Pero ya lo tenemos preparado para contestar la pregunta: ¿En qué semanas estuvo en la posición 1 cada una de las canciones de Rock que ingresaron antes de la segunda mitad del año 2000?

```python
resultado = billboard_tidy.loc[(billboard_tidy['genre']=='Rock')&(billboard_tidy['date.entered']<='2000-06-30')&(billboard_tidy['ranking']==1), ['artist.inverted', 'track', 'week']]
resultado
```

```python
# Podemos mejorar el resultado ordenando el resultado con la función sort_values
resultado.sort_values(['track', 'week'])
```
