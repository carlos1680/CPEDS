# 3. Tidy data

Recordemos que las condiciones para que un dataset se considere ordenado (tidy) son:

* Cada variable forma ______
* Cada observación forma _______
* Cada tipo de unidad observacional forma _________

¿El dataset de notas cumple con estas condiciones?

```python
# Creamos nuestro dataset tidy
notas_tidy = pd.melt(notas,
                        id_vars = [	'ID', 'Título', 'Universidad'],
                        var_name='Trabajo',
                        value_name='Nota')
notas_tidy
```

```python
# Dimensiones del dataset
notas_tidy.shape
```

Ahora podemos contestar la pregunta:

¿En cuáles trabajos prácticos cada persona se saco una nota igual o superior a 95?

```python
condicion = notas_tidy['Nota']>=95
notas_tidy.loc[condicion, ['ID','Trabajo','Nota']].sort_values(by=['ID','Trabajo','Nota'],ascending=[True,True,False])
```

## Opcionales

### ¿Cuantos trabajos prácticos de cada tipo tuvieron una nota igual o superior a 95 puntos? 

```python
condicion = notas_tidy['Nota'] >= 95
notas_tidy.loc[condicion, ['Trabajo']].groupby('Trabajo', group_keys=False).agg(
    {
        'Trabajo':[("Cantidad","count")]
    }
)
```

### ¿Cuál fue la nota promedio por cada trabajo práctico? 

```python
notas_tidy.loc[condicion, ['Trabajo','Nota']].groupby('Trabajo').agg(
    {
        'Nota': [("Promedio", "mean")]
    }
).sort_values(by='Nota',ascending = False)
```
