## 2.1 Distancia euclidea

Para ello vamos a utilizar la distancia euclidea entre los puntos.

La fórmula para calcular la distancia euclidea entre dos puntos $(x_1, y_1)$ y $(x_2, y_2)$es:

$d=\sqrt{(x_1 - x_2)^2+(y_1 - y_2)^2}$
