## 2.1 - Dividiendo un dataset

```python
# Importamos el dataset de entrenamiento y lo dividimos en dos iguales
url = '../DataSets/house_train.csv'
df = pd.read_csv(url)
#train = pd.read_csv("./sample_data/house_train.csv")
```

```python
#Antes de empezar, unas consideraciones sobre head, head(x) y tail 
df.head()
```

```python
len(df)
```

```python
# partimos en partes iguales
df_1 = df.loc[:len(df)/2-1,:]
df_2 = df.loc[len(df)/2:]

print(len(df_1), len(df_2))
```

```python
df_1.tail()
```

```python
df_2.head()
```

```python
# particion no concecutiva
df_1 = df.sample(frac = 0.5, random_state = 1234)
df_2 = df[~df.Id.isin(df_1.Id)]

print(len(df_1), len(df_2))
```
