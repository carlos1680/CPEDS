
# 3 - Joins




##3.1 - *inner join*

Vamos a unir los datasets *week1* y *week2* seleccionando el tipo de union que queremos realizar, Primero vamos a hacer el join por *Customer_ID*.

```python
ventas_merge = semana1.merge(semana2, 
                             how = "inner", 
                             on = "Customer_ID", 
                             suffixes = [" - A", " - B"])

ventas_merge.head()
```

```python
pd.merge(left = semana1, 
         right = semana2, 
         left_on = "Customer_ID",
         right_on = "Customer_ID",
         how = "inner", 
         #on = "Customer_ID", 
         suffixes = [" - A", " - B"])
```

## 3.2 - *outer join*


```python
ventas_o = semana1.merge(semana2, 
                         how = "outer", 
                         on = "Customer_ID", 
                         suffixes = [" - Week 1", " - Week 2"],
                         indicator = True)

ventas_o.head()
```

```python
print("Quedaron {} filas en vez de 500 ... ¿Por qué?".format(len(ventas_o)))
```

```python
ventas_o["_merge"].value_counts()
```

```python
# Si queremos ver las que estan en un solo datasets
ventas_o[ventas_o["_merge"].isin(["left_only", "right_only"])]
```

```python
#Hay un desfasaje entre el Indice y el ID, por lo tanto si solo indicamos la parte izquierda del merge 
clientes.query('ID == 941')
```

```python
ventas_cliente = semana1.merge(clientes, 
                               how = "left", 
                               left_on = "Customer_ID", 
                               right_index = True)
ventas_cliente.head()
```

```python
ventas_cliente.query('Customer_ID == 941')
```

```python
ventas_cliente_2 = semana1.merge(clientes, 
                                 how = "left", 
                                 left_on = "Customer_ID", 
                                 right_on = "ID")
ventas_cliente_2.head()
```

```python
#Ahora si tenemos el cliente correcto, pero nos sobra una columna!
ventas_cliente_2.query('Customer_ID == 941')
```

```python
ventas_cliente_2 = ventas_cliente_2.drop(['ID'], axis = 1)
ventas_cliente_2.head()
```

```python
ventas_final = ventas_cliente_2.merge(comidas, 
                                      how = "left", 
                                      left_on = "Food_ID", 
                                      right_on = "Food_ID")
ventas_final.head()
```
