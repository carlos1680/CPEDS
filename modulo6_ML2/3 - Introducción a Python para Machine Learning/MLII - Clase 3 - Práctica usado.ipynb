### 쮺칩mo funciona Lasso?

- Penalizaci칩n L1: Lasso introduce la penalizaci칩n en la forma de la norma L1 de los coeficientes de las variables independientes. Esto fuerza a algunos coeficientes a hacerse exactamente cero, lo que tiene el efecto de seleccionar caracter칤sticas importantes y descartar las menos relevantes.
    
    
- Selecci칩n de variables: Una de las propiedades m치s importantes de Lasso es que realiza una selecci칩n autom치tica de variables. Las caracter칤sticas con coeficientes que se vuelven cero son efectivamente eliminadas del modelo. Esto simplifica el modelo al enfocarse solo en las caracter칤sticas que realmente importan.
    
- Reducir el overfitting: Al penalizar las variables irrelevantes y reducir los coeficientes grandes, Lasso ayuda a crear un modelo m치s simple y generalizable.

GridSearchCV es una t칠cnica de b칰squeda exhaustiva utilizada en Machine Learning para encontrar los mejores hiperpar치metros para un modelo. Es especialmente 칰til cuando est치s trabajando con modelos que tienen hiperpar치metros que afectan su rendimiento, y deseas encontrar la combinaci칩n que optimice los resultados.

En Machine Learning, los hiperpar치metros son par치metros que no se aprenden directamente del modelo, sino que deben ser establecidos antes del proceso de entrenamiento.

### 쮺칩mo funciona GridSearchCV?

Especificar el modelo y los hiperpar치metros:

Seleccionas el modelo que deseas ajustar (como un 치rbol de decisi칩n, regresi칩n log칤stica, etc.).
Luego, defines un conjunto de hiperpar치metros que deseas probar. Cada hiperpar치metro tendr치 un conjunto de valores posibles  (como un rango de valores de regularizaci칩n, el n칰mero de vecinos para k-NN, etc.).

B칰squeda en la cuadr칤cula (grid):

GridSearchCV crea una "cuadr칤cula" de todas las combinaciones posibles de los hiperpar치metros que has especificado.
Por ejemplo, si tienes dos hiperpar치metros con 3 valores posibles cada uno, se probar치n 9 combinaciones en total.
Entrenamiento y validaci칩n cruzada:

Para cada combinaci칩n de hiperpar치metros, GridSearchCV entrena el modelo utilizando validaci칩n cruzada, dividendo el conjunto de datos en varias particiones (folds).

El modelo se entrena en una parte y se valida en la otra, repitiendo esto para todas las particiones.

Evaluaci칩n del rendimiento:

Despu칠s de probar todas las combinaciones, GridSearchCV eval칰a el rendimiento del modelo (por ejemplo, utilizando precisi칩n, F1-score, o cualquier m칠trica relevante) y selecciona la mejor combinaci칩n de hiperpar치metros basada en el rendimiento promedio en las validaciones cruzadas.

Mejores hiperpar치metros:

Una vez que se completa la b칰squeda, GridSearchCV te dice cu치l es la mejor combinaci칩n de hiperpar치metros que maximiza el rendimiento del modelo.

- Linealidad

Qu칠 significa: La relaci칩n entre las variables independientes y la variable dependiente debe ser lineal. Esto quiere decir que si trazas una l칤nea recta a trav칠s de los puntos de datos, debe ajustarse bien.
Por qu칠 es importante: Si la relaci칩n no es lineal (por ejemplo, si los datos forman una curva), la regresi칩n lineal no ser치 capaz de capturar esa forma, y los resultados del modelo ser치n incorrectos.

```python
# 1. Verificaci칩n de no linealidad: Gr치fico de residuos vs valores predichos
plt.figure(figsize=(6,4))
plt.scatter(y_pred, residuals)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Valores Predichos')
plt.ylabel('Residuos')
plt.title('Residuos vs Valores Predichos')
plt.show()
```

La predicci칩n realizada por el modelo est치 en el eje X y la precisi칩n de la predicci칩n est치 en el eje Y.

Los valores positivos para el residual (en el eje Y) significan que la predicci칩n era demasiado baja, y los valores negativos significan que la predicci칩n era demasiado alta; 0 significa que la estimaci칩n era exactamente correcta.


Los residuos deben estar distribuidos aleatoriamente alrededor de la l칤nea horizontal en 0 en este tipo de gr치fico.
Patr칩n observable en el gr치fico: En este gr치fico, los residuos no parecen estar distribuidos de manera aleatoria. En lugar de eso, muestran un patr칩n donde los residuos disminuyen a medida que aumentan los valores predichos.

- Homoscedasticidad (constancia de la varianza de los errores)

Qu칠 significa: Los errores deben tener una varianza constante en todos los niveles de las variables independientes. En otras palabras, la dispersi칩n de los errores deber칤a ser m치s o menos la misma en todas partes.
Por qu칠 es importante: Si los errores var칤an mucho en diferentes partes del gr치fico, significa que el modelo es inestable y podr칤a no funcionar bien para algunos datos.
Ejemplo visual: Si est치s prediciendo el precio de casas, los errores deber칤an ser parecidos tanto para casas peque침as como grandes. Si los errores son mucho mayores para las casas grandes que para las peque침as, hay heterocedasticidad (lo opuesto a homocedasticidad).

- Normalidad de los errores

Qu칠 significa: Los errores (residuos) deben seguir una distribuci칩n normal (en forma de campana), con la mayor칤a de los errores concentrados cerca de 0.
Por qu칠 es importante: Esto asegura que el modelo haga buenas predicciones para la mayor칤a de los datos y no se vea afectado por valores extremos (outliers).
Ejemplo visual: Piensa en el precio de las casas. Si el modelo predice bien la mayor칤a de las veces pero tiene algunos errores muy grandes (por ejemplo, predice 100.000 cuando el precio es 500.000), los errores no son normales.
